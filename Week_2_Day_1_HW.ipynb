Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors. Gracefully log the error if the user doesn't enter a number

```python
#adult = 18-65
#kids = <18
#seniors = >65

age = int(input('What is your age in years.'))
#if a person is under 18
if age <= 17:
    print ("kids")
#if a person is older than 17 but younger or same as 66, if a person is between 18-65
elif age > 17 and age <=65 :   
    print("adults")
elif  age >= 66:
    print ("seniors")

```
